```python
# major directory / file names
import os
import shutil
import json
import csv
import requests
import pickle
import datetime
import time


raw_wikidata_dir = 'full_data/human_wikidata_files/'
target_wikidata_dir = "target_wikidata_files/"

final_data_dir = "final_data/"
if os.path.exists(final_data_dir):
    shutil.rmtree(final_data_dir)
os.makedirs(final_data_dir)

node_list_file_path = os.path.join(final_data_dir, "node_list.csv")
edge_list_file_path = os.path.join(final_data_dir, "edge_list.csv")
detail_json_dir = os.path.join(final_data_dir, "detail_json")
```

```python
# copy target files to target_wikidata_dir

if os.path.exists(target_wikidata_dir):
    shutil.rmtree(target_wikidata_dir)
os.makedirs(target_wikidata_dir)

for filename in os.listdir(raw_wikidata_dir):
    file_name_without_extension = os.path.splitext(filename)[0]
    if file_name_without_extension in target_id_list:
        source_path = os.path.join(raw_wikidata_dir, filename)
        target_path = os.path.join(target_wikidata_dir, filename)
        shutil.copyfile(source_path, target_path)
```

```python
# generate node list, edge list

# Create empty lists to store the node and edge data
node_data = []
edge_data = set()
name_dict = {}
outgoing_edge = {}

edge_properties = {
    'P185': 'taught',  # doctralStudent
    'P802': 'taught',  # student
    'P1066': 'studentOf',  # need switch
    'P737': 'influencedBy', # need switch
    'P184': 'studentOf',  # doctralAdvisor
}

# stats
node_count = 0
no_id = 0
no_eng_name = 0
no_desc = 0
no_birth = 0
no_death = 0
no_wiki = 0
no_edge_end = 0
no_edge = 0
edge_to_phil = 0
taught = 0
influenced = 0

newly_found_property = set()

# Loop through each file in the input directory
for filename in os.listdir(target_wikidata_dir):
    # Open the file and load the JSON data
    with open(os.path.join(target_wikidata_dir, filename), "r") as f:
        data = json.load(f)
    
    # Extract the node data and append it to the node_data list

    for entity_id, entity_data in data['entities'].items():

        node_count += 1
        try:
            node_id = entity_data["id"]
        except:
            node_desc = ""
            no_id += 1
            continue

        try:
            node_name = entity_data["labels"]["en"]["value"]
            name_dict[node_id] = node_name
        except:
            try:
                node_name = entity_data["aliases"]["en"]["value"]
                name_dict[node_id] = node_name
            except:
                no_eng_name += 1
                try:
                    node_name = entity_data["labels"].popitem()[1]["value"]
                    name_dict[node_id] = node_name
                except:
                    node_name = ""
                    continue

        try:
            node_desc = entity_data["descriptions"]["en"]["value"]
        except:
            node_desc = ""
            no_desc += 1

        try:
            node_birth = entity_data["claims"]["P569"][0]["mainsnak"]["datavalue"]["value"]["time"]
        except:
            node_birth = ""
            no_birth += 1

        try:
            node_death = entity_data["claims"]["P570"][0]["mainsnak"]["datavalue"]["value"]["time"]
        except:
            node_death = ""
            no_death += 1
        
        try:
            wiki_link = entity_data["sitelinks"]["enwiki"]["url"]
        except:
            wiki_link = ""
            no_wiki += 1

        node_data.append([node_id, node_name, node_desc, node_birth, node_death, wiki_link])
        this_has_edge = False 
        if (outgoing_edge.get(node_id) == None):
            outgoing_edge[node_id] = 0

        for property_id, claims in entity_data['claims'].items():
            for claim in claims:
                edge_to_node = False
                try:
                    if claim['mainsnak']['snaktype'] == "somevalue" or claim['mainsnak']['snaktype'] == "novalue":
                        continue
                    if claim['mainsnak']['datavalue']['type'] == "wikibase-entityid":
                        value = claim['mainsnak']['datavalue']['value']['id']
                        if value in target_id_list:
                            edge_to_node = True
                            edge_to_phil += 1
                            edge_type = edge_properties[property_id]

                            from_id = node_id
                            to_id = value

                            if (edge_type == 'taught'):
                                taught += 1
                            if (edge_type == 'influencedBy'):
                                edge_type = 'influenced'
                                influenced += 1
                                from_id, to_id = to_id, from_id
                            if (edge_type == 'studentOf'): 
                                edge_type = 'taught'
                                taught += 1
                                from_id, to_id = to_id, from_id

                            if (outgoing_edge.get(from_id) == None):
                                outgoing_edge[from_id] = 1
                            else:
                                outgoing_edge[from_id] += 1       
                except:
                    edge_to_node = False
                    value = ""
                    print(f"error in {node_id} {property_id}")
                    newly_found_property.add(property_id)
                    no_edge_end += 1

                if (edge_to_node):
                    edge_data.add((from_id, edge_type, to_id))
        
        if this_has_edge == False:
            no_edge += 1

have_outgoing_edge = 0
for node, count in sorted(outgoing_edge.items(), 
                            key=lambda item: item[1],
                            reverse=True):
    if (count > 0):
        have_outgoing_edge += 1
    if (count > 3):
        print(f"{count} outgoing edges: \t {name_dict[node]}({node})")

print(f"\n\n\nProcessed files in {target_wikidata_dir}\n")
print(f"no id = {no_id}/{(no_id*100/node_count):.2f}%, no eng name = {no_eng_name}/{no_eng_name*100/node_count:.2f}%, no desc = {no_desc}/{no_desc*100/node_count:.2f}%,")
print(f"no birth = {no_birth}/{no_birth*100/node_count:.2f}%, no death {no_death}/{no_death*100/node_count:.2f}%, no wiki = {no_wiki}/{no_wiki*100/node_count:.2f}%")
print(f"have outgoing influence = {have_outgoing_edge}/{have_outgoing_edge*100/node_count:.2f}%")
print(f"edge to another philosopher = {edge_to_phil}")
print(f"edge type = taught = {taught}/{taught*100/edge_to_phil:.2f}%, influenced = {influenced}/{influenced*100/edge_to_phil:.2f}%")
print(f"newly found property = {newly_found_property}")


# Write the node data to the node_output_file
with open(node_list_file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["id", "name", "node_desc", "node_birth", "node_death", "wiki_link"])
    writer.writerows(node_data)

# Write the edge data to the edge_output_file
with open(edge_list_file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["start_node", "edge_type", "end_node"])
    writer.writerows(edge_data)
```

```python
# generate detail.json

# New dictionary with keys as P#### and values as natural languages
properties = {
    'P569': 'dateOfBirth',
    'P570': 'dateOfDeath',
    'P2348': 'timePeriod',
    'P69': 'educatedAt',
    'P27': 'countryOfCitizenship',
    'P551': 'residence',
    'P625': 'coordinates',
    'P108': 'employer',
    'P131': 'locatedIn',
    'P106': 'occupation',
    #'P2959': 'permanentDuplicatedItem',
    'P101': 'fieldOfWork',
    'P135': 'movement',
    'P185': 'doctralStudent',
    'P802': 'student',
    'P1066': 'studentOf',
    'P737': 'influencedBy',
    'P18': 'image',
    'P800': 'notableWork'
}

# Dictionary to cache fetched Q### names
cached_names = {}
cached_coords = {}
claim_counter = 0
hit_count = 0
start_time = time.time()

def find_name(Q_id):
    global claim_counter
    global hit_count
    if (claim_counter == 0):
        claim_counter += 1
    if Q_id in cached_names:
        hit_count += 1
        return cached_names[Q_id]

    url = f"https://www.wikidata.org/wiki/Special:EntityData/{Q_id}.json"

    try:
        entity_data = requests.get(url).json()['entities'][Q_id]
        node_name = entity_data["labels"]["en"]["value"]
    except:
        try:
            node_name = entity_data["aliases"]["en"]["value"]
        except:
            try:
                node_name = entity_data["labels"].popitem()[1]["value"]
            except:
                node_name = ""
    
    cached_names[Q_id] = node_name
    return node_name

# Load cached_names and cached_coords dictionaries

cache_file_path = 'cached_data_2.pkl'
if os.path.exists(cache_file_path):
    with open(cache_file_path, 'rb') as file:
        cached_names, cached_coords = pickle.load(file)

# Initialize a dictionary to store the JSON objects
json_objects = {}


# Read the node list CSV file
with open(node_list_file_path, "r") as file:

    reader = csv.reader(file)
    next(reader)  # Skip the header row

    total_rows = sum(1 for _ in reader)
    file.seek(0)
    next(reader)  # Skip the header row
    counter = 0
    
    for row in reader:
        counter += 1
        node_id = row[0]
        json_objects[node_id] = {"id": node_id, "name": row[1], "desc": row[2], "wiki_link": row[5]}
        json_file_path = os.path.join(target_wikidata_dir, f"{node_id}.json")
        with open(json_file_path, "r") as json_file:
            data = json.load(json_file)

            for property_id, claims in data['entities'][node_id]['claims'].items():
                if property_id not in properties:
                    continue
                key = properties[property_id]
                for claim in claims:
                    if claim['mainsnak']['snaktype'] == "somevalue" or claim['mainsnak']['snaktype'] == "novalue":
                        continue
                    if claim['mainsnak']['datavalue']['type'] == "wikibase-entityid":
                        if (claim_counter > 0):
                            claim_counter += 1
                        value = claim['mainsnak']['datavalue']['value']['id']
                        coordinates = "none"
                        if key == 'residence':
                            if value not in cached_coords:
                                url = f"https://www.wikidata.org/wiki/Special:EntityData/{value}.json"
                                try:
                                    entity_data = requests.get(url).json()['entities'][value]
                                    coordinates = entity_data["claims"]["P625"][0]["mainsnak"]["datavalue"]["value"]
                                except:
                                    coordinates = ""
                                cached_coords[value] = coordinates
                            else:
                                coordinates = cached_coords[value]
                        elif key == 'employer' or key == 'educatedAt':
                            if value not in cached_coords:
                                url = f"https://www.wikidata.org/wiki/Special:EntityData/{value}.json"
                                try:
                                    entity_data = requests.get(url).json()['entities'][value]
                                    city = entity_data["claims"]["P276"][0]["mainsnak"]["datavalue"]["value"]
                                except:
                                    try:
                                        city = entity_data["claims"]["P131"][0]["mainsnak"]["datavalue"]["value"]
                                    except:
                                        city = value
                                url = f"https://www.wikidata.org/wiki/Special:EntityData/{value}.json"
                                try:
                                    city_entity = requests.get(url).json()['entities'][value]
                                    coordinates = city_entity["claims"]["P625"][0]["mainsnak"]["datavalue"]["value"]
                                except:
                                    try:
                                        coordinates = entity_data["claims"]["P625"][0]["mainsnak"]["datavalue"]["value"]
                                    except:
                                        coordinates = ""
                                cached_coords[value] = coordinates
                            else:
                                coordinates = cached_coords[value]

                        if value.startswith('Q'):
                            value = find_name(value)

                        if coordinates != "none":
                            value = {"name": value, "coordinates": coordinates}
                    elif claim['mainsnak']['datavalue']['type'] == "time":
                        value = claim['mainsnak']['datavalue']['value']['time']
                    else:
                        value = claim['mainsnak']['datavalue']['value']

                    # Check if the key already exists in the JSON object
                    if key in json_objects[node_id]:
                        # Get the current index of the key
                        current_index = len(json_objects[node_id][key])
                        # Append the value to the existing key with the next index
                        json_objects[node_id][key][current_index] = value
                    else:
                        # Create a new key with the value as the first element in a list
                        json_objects[node_id][key] = {0: value}
                hit_rate = 0
                if claim_counter > 0:
                    hit_rate = hit_count/claim_counter * 100

                if claim_counter % 100 == 0:
                    # Save the cached_names and cached_coords dictionaries
                    with open("cached_data_2.pkl", 'wb') as file:
                        pickle.dump([cached_names, cached_coords], file)
                
            estimated_remaining_time = int((time.time() - start_time) * (total_rows - counter) / counter)

            remaining_time = datetime.timedelta(seconds=estimated_remaining_time)
            formatted_remaining_time = str(remaining_time)
                
            print(f"Progress: {counter}/{total_rows} ({counter/total_rows*100:.2f}%) \t current hit rate = {hit_rate:.2f}% ({hit_count}/{claim_counter}) \t estimated remaining time = {formatted_remaining_time}")


    with open("cached_data_2.pkl", 'wb') as file:
        pickle.dump([cached_names, cached_coords], file) 
        
           
    if os.path.exists(detail_json_dir):
        shutil.rmtree(detail_json_dir)
    os.makedirs(detail_json_dir)

    # Save the JSON objects to separate files
    for json_file_name, json_obj in json_objects.items():
        with open(os.path.join(detail_json_dir , json_file_name + '.json'), 'w') as json_file:
            json.dump(json_obj, json_file)



```
